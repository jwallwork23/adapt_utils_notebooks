```python
from firedrake import *
from firedrake_adjoint import *
```

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
a = Constant(1.0)
b = Constant(-1.0)
```

```python
mesh = UnitSquareMesh(1, 1)
dx = dx(domain=mesh)
```

```python
J = assemble(a*dx)
```

```python
J = J + assemble(b*dx)
```

```python
tape = get_working_tape()
```

```python
G = tape.create_graph()
```

```python
fig, axes = plt.subplots(figsize=(16, 5))
options = {
    "font_size": 14,
    "node_size": 200,
    "edgecolors": "black",
    "linewidths": 2,
    "width": 2,
}
pos = nx.spring_layout(G, seed=1)  # Seed layout for reproducibility
nx.draw_networkx(G, pos=pos, ax=axes, **options)
axes.axis(False);
```
